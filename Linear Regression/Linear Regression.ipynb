# Linear Regression for comparing sand granule size and slope of beach


This notebook is used for learning,applying and understanding the basics of Linear Regression using a simple dataset which has a relation between grain size and slope of the beach . 
The data set used is https://college.cengage.com/mathematics/brase/understandable_statistics/7e/students/datasets/slr/frames/frame.ht

Importing the necessary packages and dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random 
import math
```

Reading the dataset and converting it to a numpy array

```python
df = pd.read_csv("C:/Users/Yash Phadke/OneDrive/Desktop/ML/Linear Regression/sand vs slope.csv")
array = df.to_numpy()
print(df.head())
```

#### For without bias

Plotting the dataset and assuming a best fit line

```python
slope = array.T[1]
diameter = array.T[0]
plt.scatter(slope,diameter)
x = np.linspace(0,450,1000)
y = 0.2666666667*x
plt.plot(x,y,color="black")
plt.xlabel("Diameter of sand granules")
plt.ylabel("Slope of beach")
plt.title("Diameter of sand granules vs Slope of beach")
plt.show
```

Defining the cost function 

```python
def Cost_function(weights):
    n = 0
    cost_final = 0
    while n < 63:
        cost = (array[n,0] -weights*array[n,1])**2
        cost_final = cost_final + cost
        n = n+1
    cost_function = cost_final/(2*62)
    return cost_function
Cost_function(0.2666666667)
```

Plotting the cost function with respect to its respective weight

```python
i = 0 
w =np.random.uniform(low=0.0, high=0.5, size=(1000,))
cost_y_axis = []
while i < 1000:
    temp = Cost_function(w[i])
    cost_y_axis.append(temp)
    i = i +1
```

Training Model using Gradient Descent

```python
def Diff_Cost(w):
    t = 0
    temp_weights:float64 = 0
    while t<63:
        diff = ((w*array[t,1]) - array[t,0])*array[t,1]
        temp_weights = temp_weights + diff
        t = t + 1
    diff_cost = temp_weights/62
    return diff_cost
def Grad_desc(w,alpha):
    n = 0
    while n < 10000:
        w = w - alpha*Diff_Cost(w)
        n = n + 1
    return w 
Diff_Cost(0.2666666667)
w = Grad_desc(0.2666666667,0.0001)
print(w)
```

```python
slope = array.T[1]
diameter = array.T[0]
plt.scatter(slope,diameter)
x = np.linspace(0,450,1000)
y = 0.2666666667*x
z = 0.238018143738856698*x
plt.plot(x,z,color = "red")
plt.plot(x,y,color="black")
plt.xlabel("Diameter of sand granules")
plt.ylabel("Slope of beach")
plt.title("Assumed(black) vs Best-Fit(red)")
plt.show
```

```python
plt.scatter(w,cost_y_axis,marker=".")
plt.xlabel("Weight")
plt.ylabel("Cost Function Value")
plt.title("Cost Function Value vs Weight")
plt.show()
```

#### With bias


Plotting the data and assuming the best fit line

```python
slope = array.T[1]
diameter = array.T[0]
plt.scatter(slope,diameter)
x = np.linspace(0,450,1000)
y = 0.3*x -10
plt.plot(x,y,color="black")
plt.xlabel("Diameter of sand granules")
plt.ylabel("Slope of beach")
plt.title("Diameter of sand granules vs Slope of beach")
plt.show
```

Defining Cost function

```python
def Cost_function(weights,bias):
    n = 0
    cost_final = 0
    while n< 63:
        cost = (array[n,0] -(weights*array[n,1]+bias))**2
        cost_final = cost_final + cost
        n = n+1
    cost_function = cost_final/(2*62)
    return cost_function
Cost_function(0.2666666667,-10)
```

Plotting the surface using trisurf

```python
i = 0 
w =np.random.uniform(low=0.0, high=0.8, size=(15000,))
b =np.random.uniform(low=-20, high=20, size=(15000,))
cost_y_axis = []
while i < 15000:
    temp = Cost_function(w[i],b[i])
    cost_y_axis.append(temp)
    i = i +1
```

```python
def Diff_Cost(w,b):
    t = 0
    temp_weights:float64 = 0
    while t<63:
        diff = ((w*array[t,1]+b) - array[t,0])*array[t,1]
        temp_weights = temp_weights + diff
        t = t + 1
    diff_cost = temp_weights/62
    return diff_cost
def Diff_Cost_wrtb(w,b):
    t = 0
    temp_bias:float64 = 0
    while t<63:
        diff = ((w*array[t,1]+b) - array[t,0])
        temp_bias = temp_bias + diff
        t = t + 1
    diff_cost_wrtb = temp_bias/62
    return diff_cost_wrtb
def Grad_desc(w,b,alpha):
    n = 0
    while n < 800000:
        temp_w = w - alpha*Diff_Cost(w,b)
        temp_b = b - alpha*Diff_Cost_wrtb(w,b)
        w = temp_w
        b = temp_b
        n = n + 1
    return w,b 
Diff_Cost(0.2666666667,-10)
Diff_Cost_wrtb(0.2666666667,-10)
w = Grad_desc(0.2666666667,-10,0.0001)
print(w)
```

```python
ax = plt.axes(projection="3d")
fig = plt.figure(figsize =(16, 9)) 
ax.plot_trisurf(np.array(w),np.array(b),np.array(cost_y_axis),linewidth=0.2, antialiased=True)
ax.set_xlabel('Weights', fontweight ='bold')
ax.set_ylabel('Biases', fontweight ='bold')
ax.set_zlabel('Cost Function', fontweight ='bold')
ax.set_title('Surface Of Cost function with Weights and Biases')
```
