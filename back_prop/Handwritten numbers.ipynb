Defining the activation function

```python
def sig(x):
    g = 1/(1 + np.exp(-x))
    return g
```

Here we define the function for making the layers in a NN . The inputs for the function are 
- number of neurons 
- input activation 
- weights 
- bias

```python
def layer(n,a_i,w,b):
    j = 0 
    a_t =[]
    for j in range(n):
        a = np.dot(w[j],a_i) + b[j]
        a_t.append(a)
    a_o = np.array(a_t)
    return a_o
```

Here we define the structure of the NN which will contain 2 hidden layers each of 16 neurons and the output layer having 10 neurons each one for a number 

```python
def structure(array):
    w_1 = np.random.uniform(-0.5,0.5,(16,len(array)))
    b_1 = np.random.uniform(-0.5,0.5,16)
    a_1 = layer(16,array,w_1,b_1)
    w_2 = np.random.uniform(-0.5,0.5,(16,len(a_1)))
    b_2 = np.random.uniform(-0.5,0.5,16)
    a_2 = layer(16,sig(a_1),w_2,b_2)
    w_3 = np.random.uniform(-0.5,0.5,(10,len(a_2)))
    b_3 = np.random.uniform(-0.5,0.5,10)
    a_3 = layer(10,sig(a_2),w_3,b_3)
    return sig(a_3),w_1,w_2,w_3,b_1,b_2,b_3
```

```python
final,w_1,w_2,w_3,b_1,b_2,b_3 =structure(test_np[0])
warnings.filterwarnings('ignore')
print(final)
```
